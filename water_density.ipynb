```
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_html("https://www.internetchemistry.com/chemical-data/water-density-table.php")[0]

T, rho = [], []

#print(df.columns[1:])

for i, row in df.iterrows():
  _T = row[df.columns[0]] + [float(item) for item in df.columns[1:]]
  _rho = row[df.columns[1:]].values
  T.extend(_T)
  rho.extend(1000*_rho)

##plt.plot(T, rho, "r.-")
##plt.show()

for i in range(len(T)):
  print(f"{T[i]:.1f} {rho[i]:.2f}")
```
